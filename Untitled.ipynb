# univariate selection

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
data=pd.read_csv("train.csv")
```

```python
data.head()
```

```python
X= data.iloc[: ,0:20]
y=data.iloc[:,-1]
```

```python
# apply selctk best ,top 10 features
best_features =SelectKBest(score_func=chi2,k=10)
```

```python
fit =best_features.fit(X,y)
```

```python
fit.scores_
```

```python
dfscores=pd.DataFrame(fit.scores_)
dfcolumns =pd.DataFrame(X.columns)
```

```python
fetauresscores=pd.concat([dfcolumns,dfscores],axis=1)
```

```python
fetauresscores.columns=["features","score"]
```

```python
fetauresscores.sort_values(by='score',ascending=False).head(10)
```

# Features importance

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python
model =RandomForestClassifier()
model.fit(X,y)
```

```python
model.feature_importances_   #just like the value are best on percentage ...i.e ,the 0.07683 % of the 1st features contributing 
```

```python
featuresimportance=pd.DataFrame(model.feature_importances_ ,index=X.columns,columns=["importance"])
```

```python
featuresimportance =featuresimportance.sort_values(by="importance",ascending=False)
```

```python
featuresimportance
```

```python
plt.figure(figsize=(20,6))
plt.bar(featuresimportance.index,featuresimportance['importance'])
plt.show()
```
